```python
from sklearn import linear_model
```

```python
X = []
y = []

imu_min = 5

mu_edges = np.linspace(0,1,121)
mus = [(mu_edges[row]+mu_edges[row+1])*0.5 for row in range(len(mu_edges)-1)]


for nowid in range(1): 
    redshift = 1.5 - nowid * 0.1
    CITA_mu_DATAnowid = CITA_DATA[nowid].reshape(150, 120, 7)
    xi_mu_CITA[nowid] = CITA_mu_DATAnowid[5:40,:,2].sum(0)[::-1]
    
    ### gui yi hua... xi_mu
    
    ### add samples into X, y...
    
    for imu in range(len(mus)):
        if imu < imu_min: continue
        
        ### add redshift high order terms... redshift * mu term...
        X.append([redshift, mus[imu], mus[imu]**2.0, mus[imu]**3.0,])
                 #mus[imu]**9.0])
        y.append(xi_mu_CITA[nowid][imu])
    
X = np.array(X)
y = np.array(y)
```

### fit a model 

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X,y)
y_predict = model.predict(X)
```
