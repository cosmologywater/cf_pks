```python
def draws(nowid):
    fig = plt.figure(figsize=(12,8))
    ax = fig.add_subplot(111)
    plt.title('s-xi(s)',fontdict={'family' : 'Times New Roman', 'size'   : 30},y=1.05)

    smax = 40
    smin = 5
    z = 1.5-nowid*0.1
    
    CITA_normnowid = [float(x) for x in (open(str(CITA[nowid]),'r').readline().split())[1:4]]
    COLA_CITA_normnowid = [float(x) for x in (open(str(COLA_CITA[nowid]),'r').readline().split())[1:4]]


    CITA_DATA[nowid] = CITA_DATA[nowid].reshape(150, 120, 7)
    DD_CITA_nowid, RD_CITA_nowid, RR_CITA_nowid  = CITA_DATA[nowid][smin:smax,:,3].sum(1), \
    CITA_DATA[nowid][smin:smax,:,4].sum(1), CITA_DATA[nowid][smin:smax,:,6].sum(1)
    xi_CITA[nowid] = ( DD_CITA_nowid/ CITA_normnowid[0] - 2*RD_CITA_nowid/CITA_normnowid[1] \
                      + RR_CITA_nowid/CITA_normnowid[2] ) / (RR_CITA_nowid/CITA_normnowid[2]) 
 
    X = np.linspace(smin+0.5,smax-0.5,smax-smin)
    ax.plot(X , meannorm(X**2 * xi_CITA[nowid]),label='CITA'+' z = '+str(z)+' 2pcf')  
    
    
    COLA_CITA_DATA[nowid] = COLA_CITA_DATA[nowid].reshape(150, 120, 7)
    DD_COLA_CITA_nowid, RD_COLA_CITA_nowid, RR_COLA_CITA_nowid  = \
    COLA_CITA_DATA[nowid][smin:smax,:,3].sum(1), COLA_CITA_DATA[nowid][smin:smax,:,4].sum(1), \
    COLA_CITA_DATA[nowid][smin:smax,:,6].sum(1)
    xi_COLA_CITA[nowid] = ( DD_COLA_CITA_nowid/ COLA_CITA_normnowid[0] - 2*RD_COLA_CITA_nowid/COLA_CITA_normnowid[1] \
                      + RR_COLA_CITA_nowid/COLA_CITA_normnowid[2] ) / (RR_COLA_CITA_nowid/COLA_CITA_normnowid[2]) 
    ax.plot( X, meannorm(X**2 * xi_COLA_CITA[nowid]),label='COLA_CITA'+' z = '+str(z)+' 2pcf',linewidth=3.0,linestyle=':')


    plt.tick_params(labelsize=20)
    labels = ax.get_xticklabels() + ax.get_yticklabels()
    [label.set_fontname('Times New Roman') for label in labels]
    ax.legend(prop={'family' : 'Times New Roman','size'   : 14,})
    plt.xlabel('s',{'family' : 'Times New Roman','size'   : 28,})
    plt.ylabel('xi(s)',{'family' : 'Times New Roman','size'   : 24,})
    ax.grid(c='white')
```

```python
X = []
y = []

imu_min = 5

mu_edges = np.linspace(0,1,121)
mus = [(mu_edges[row]+mu_edges[row+1])*0.5 for row in range(len(mu_edges)-1)]


for nowid in range(1): 
    redshift = 1.5 - nowid * 0.1
    CITA_mu_DATAnowid = CITA_DATA[nowid].reshape(150, 120, 7)
    xi_mu_CITA[nowid] = CITA_mu_DATAnowid[5:40,:,2].sum(0)[::-1]
    
    ### gui yi hua... xi_mu
    
    ### add samples into X, y...
    
    for imu in range(len(mus)):
        if imu < imu_min: continue
        
        ### add redshift high order terms... redshift * mu term...
        X.append([redshift, mus[imu], mus[imu]**2.0, mus[imu]**3.0,])
                 #mus[imu]**9.0])
        y.append(xi_mu_CITA[nowid][imu])
    
X = np.array(X)
y = np.array(y)
```
